```python
file = open('obituary_text_corpus_GPT2_5.txt','w+') 
```

```python
import requests
from bs4 import BeautifulSoup

with open('obit_links.txt') as lines: 
    for line in lines:
        next_url = ("http://oa.anu.edu.au" + line)
        resp = requests.get(next_url)
        #print(next_url)
        html_str = resp.text
        
        document = BeautifulSoup(html_str, "html.parser")
        type(document)
        
        h2_tag = document.find('h2')
        h2_tag.string
        
        file.write(h2_tag.string)
        file.write(" ")
        
        for para in document.select("div.obituaryContent p"):
            if(para.string):
                file.write(para.string)
                file.write("\@")
```
